### Scraping the Data

```python
# empty lists to store scraped data temporarily
scraped_names = []
scraped_years = []
scraped_wins = []
scraped_losses = []
scraped_ot_losses = []
scraped_win_pct = []
scraped_gfs = []
scraped_gas = []
scraped_ratings = []
```

```python
page_no = 1
while page_no <= 24:
    url = f'https://www.scrapethissite.com/pages/forms/?page_num={page_no}'
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/129.0.0.0 Safari/537.36"}
    page = requests.get(url, headers=headers)
    # print(page)

    soup = BeautifulSoup(page.text, 'html.parser')
    print(page_no)
    
    # scrape team names
    scraped_names = soup.find_all('td', class_='name')
    scraped_names = [name.text.strip() for name in scraped_names]
    names.extend(scraped_names)
    scraped_names.clear()
    # print(scraped_names)

    # scrape the years
    scraped_years = soup.find_all('td', class_='year')
    scraped_years = [year.text.strip() for year in scraped_years]
    years.extend(scraped_years)
    scraped_years.clear()
    # print(scraped_years)

    # scrape the wins
    scraped_wins = soup.find_all('td', class_='wins')
    scraped_wins = [win.text.strip() for win in scraped_wins]
    wins.extend(scraped_wins)
    scraped_wins.clear()

    # scrape the losses
    scraped_losses = soup.find_all('td', class_='losses')
    scraped_losses = [loss.text.strip() for loss in scraped_losses]
    losses.extend(scraped_losses)
    scraped_losses.clear()

    # scrape the ot-losses
    scraped_ot_losses = soup.find_all('td', class_='ot-losses')
    scraped_ot_losses = [ot_loss.text.strip() for ot_loss in scraped_ot_losses]
    ot_losses.extend(scraped_ot_losses)
    scraped_ot_losses.clear()

    # scrape the win %
    scraped_win_pct = soup.find_all('td', class_='pct')
    scpaped_win_pct = [pct.text.strip() for pct in scraped_win_pct]
    win_pct.extend(scpaped_win_pct)
    scraped_win_pct.clear()

    # scrape the goals for (GF)
    scraped_gfs = soup.find_all('td', class_='gf')
    scraped_gfs = [gf.text.strip() for gf in scraped_gfs]
    gfs.extend(scraped_gfs)
    scraped_gfs.clear()

    # scrape the goals against (GA)
    scraped_gas = soup.find_all('td', class_='ga')
    scraped_gas = [ga.text.strip() for ga in scraped_gas]
    gas.extend(scraped_gas)
    scraped_gas.clear()

    # scrape the ratings
    scraped_ratings = soup.find_all('td', class_='diff')
    scraped_ratings = [rating.text.strip() for rating in scraped_ratings]
    ratings.extend(scraped_ratings)
    scraped_ratings.clear()

    # increment the page no. to move to the next page
    page_no += 1
```

### Bit of Data Cleaning and Changing Datatypes to Appropriate Ones
